```python
#Grabcut Foreground Extraction  ##working
import cv2
import numpy as np
import matplotlib.pyplot as plt

img=cv2.imread("C:\\Users\\91910\\Downloads\\watch.jpg")
mask=np.zeros(img.shape[:2],np.uint8)
bdgModel=np.zeros((1,65),np.float64)
fdgModel=np.zeros((1,65),np.float64)
rect=(50,50,800,500)
cv2.grabCut(img,mask,rect,bdgModel,fdgModel,5,cv2.GC_INIT_WITH_RECT)
mask2=np.where((mask==2)|(mask==0),0,1).astype('uint8')
img=img*mask2[:,:,np.newaxis]
plt.imshow(img)
plt.colorbar()
plt.show()
```

```python
#Brute force matching  ##working
import cv2
import numpy as np
import matplotlib.pyplot as plt

img1=cv2.imread("C:\\Users\\91910\\Downloads\\booktemplate.jpeg",1)
img2=cv2.imread("C:\\Users\\91910\\Downloads\\book.jpeg",1)
orb=cv2.ORB_create()
kp1,des1=orb.detectAndCompute(img1,None)
kp2,des2=orb.detectAndCompute(img2,None)
bf=cv2.BFMatcher(cv2.NORM_HAMMING,crossCheck=True)
matches=bf.match(des1,des2)
matches=sorted(matches,key=lambda x:x.distance)
img3= cv2.drawMatches(img1,kp1,img2,kp2,matches[:10],None,flags=2) #we drew 10 matches
plt.imshow(img3)
plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#addition of logo onto an image  ##working

import cv2
import numpy as np
import matplotlib.pyplot as plt

img1 = cv2.imread('C:\\Users\\91910\\Downloads\\watch.jpg')
img2 = cv2.imread('C:\\Users\\91910\\Downloads\\mainlogo.jpeg')
rows,cols,channels = img2.shape
roi = img1[0:rows, 0:cols ]
img2gray = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
ret, mask = cv2.threshold(img2gray, 220, 255, cv2.THRESH_BINARY_INV)
mask_inv = cv2.bitwise_not(mask)
img1_bg = cv2.bitwise_and(roi,roi,mask = mask_inv)
img2_fg = cv2.bitwise_and(img2,img2,mask = mask)
dst = cv2.add(img1_bg,img2_fg)
img1[0:rows, 0:cols ] = dst
cv2.imshow('res',img1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#color filtering ##working

import cv2
import numpy as np
cap = cv2.VideoCapture(0)
while(1):
    _, frame = cap.read()
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    #dark_red = np.uint8([[[12,22,121]]])
    #dark_red = cv2.cvtColor(dark_red,cv2.COLOR_BGR2HSV)
    lower_red = np.array([30,150,50])
    upper_red = np.array([255,255,180])
    mask = cv2.inRange(hsv, lower_red, upper_red)
    res = cv2.bitwise_and(frame,frame, mask= mask)
    cv2.imshow('frame',frame)
    cv2.imshow('mask',mask)
    cv2.imshow('res',res)
    k = cv2.waitKey(5) & 0xFF
    if k == 27:
        break
cv2.destroyAllWindows()
cap.release()
```

```python
#blurring and smoothing  ##working

import cv2
import numpy as np
cap = cv2.VideoCapture(0)
while(1):
    _, frame = cap.read()
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    lower_red = np.array([30,150,50])
    upper_red = np.array([255,255,180])
    mask = cv2.inRange(hsv, lower_red, upper_red)
    res = cv2.bitwise_and(frame,frame, mask= mask)
    kernel = np.ones((15,15),np.float32)/225
    smoothed = cv2.filter2D(res,-1,kernel)
    cv2.imshow('Original',frame)
    cv2.imshow('Averaging',smoothed)
    k = cv2.waitKey(5) & 0xFF
    if k == 27:
        break
cv2.destroyAllWindows()
cap.release()

#gaussian blurring
#blur = cv2.GaussianBlur(res,(15,15),0)
#cv2.imshow('Gaussian Blurring',blur)

#median blurring
#median = cv2.medianBlur(res,15)
#cv2.imshow('Median Blur',median)

#bilateral blurring
#bilateral = cv2.bilateralFilter(res,15,75,75)
#cv2.imshow('bilateral Blur',bilateral)
```

```python
#morphological Transformation on opencv  ##working
#erosion and dilation
import cv2
import numpy as np

cap = cv2.VideoCapture(0)
while(1):
    _, frame = cap.read()
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    lower_red = np.array([30,150,50])
    upper_red = np.array([255,255,180])
    mask = cv2.inRange(hsv, lower_red, upper_red)
    res = cv2.bitwise_and(frame,frame, mask= mask)
    kernel = np.ones((5,5),np.uint8)
    erosion = cv2.erode(mask,kernel,iterations = 1)
    dilation = cv2.dilate(mask,kernel,iterations = 1)
    cv2.imshow('Original',frame)
    cv2.imshow('Mask',mask)
    cv2.imshow('Erosion',erosion)
    cv2.imshow('Dilation',dilation)
    k = cv2.waitKey(5) & 0xFF
    if k == 27:
        break
cv2.destroyAllWindows()
cap.release()

#removing false positives-opening
#removing false negatives-closing
#opening = cv2.morphologyEx(mask,cv2.MORPH_OPEN,kernel)
#dilation = cv2.morphologyEx(mask,cv2.MORPH_CLOSE,kernel)
#cv2.imshow('Original',frame)
#cv2.imshow('Mask',mask)
#cv2.imshow('opening,erosion)
#cv2.imshow('closing',dilation)
```

```python
#Haar Cascade Object Detection
#face and eye detection
import numpy as np
import cv2
# multiple cascades: https://github.com/Itseez/opencv/tree/master/data/haarcascades
#https://github.com/Itseez/opencv/blob/master/data/haarcascades/haarcascade_frontalface_default.xml
face_cascade = cv2.CascadeClassifier('C:\\Users\\91910\\anaconda3\\Lib\\site-packages\\cv2\\data\\haarcascade_frontalface_default.xml')
#https://github.com/Itseez/opencv/blob/master/data/haarcascades/haarcascade_eye.xml
eye_cascade = cv2.CascadeClassifier('C:\\Users\\91910\\anaconda3\\Lib\\site-packages\\cv2\\data\\haarcascade_eye.xml')
cap = cv2.VideoCapture(0)
while 1:
    ret, img = cap.read()
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for (x,y,w,h) in faces:
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]
        eyes = eye_cascade.detectMultiScale(roi_gray)
        for (ex,ey,ew,eh) in eyes:
            cv2.rectangle(roi_color,(ex,ey),(ex+ew,ey+eh),(0,255,0),2)
    cv2.imshow('img',img)
    k = cv2.waitKey(30) & 0xff
    if k == 27:
        break
cap.release()
cv2.destroyAllWindows()
```
